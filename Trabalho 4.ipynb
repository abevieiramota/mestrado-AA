# Funções

```python
def multi_scatter(dados_list, d_list, colors, legend):
    
    dims = [(x,y) for i, x in enumerate(d_list) for y in d_list[i+1:]]
    fig = pl.figure(figsize=(22, 15))

    grid_x = ceil(len(dims)/3.)
    grid_y = 3
    
    for i, (x_d, y_d) in enumerate(dims):
        
        ax = fig.add_subplot(grid_x, grid_y, i+1)
        
        for dados, color in zip(dados_list, colors):
            
            ax.scatter(dados[:, x_d], dados[:, y_d], c=color) 
            ax.set_title("features (%d, %d)" % (x_d, y_d))
            ax.legend(legend, loc=0)        
```

## Randomização e partição dos dados

```python
is_random = [np.random.permutation(np.arange(0, 30, 1)) for _ in xrange(3)]

c1_treino = np.take(c1, is_random[0], axis=0)
c1_teste = np.delete(c1, is_random[0], axis=0)

c2_treino = np.take(c2, is_random[1], axis=0)
c2_teste = np.delete(c2, is_random[1], axis=0)

c3_treino = np.take(c3, is_random[2], axis=0)
c3_teste = np.delete(c3, is_random[2], axis=0)
```

## Cálculo das médias e matrizes de covariância

```python
medias_c1 = c1_treino[:, :-1].mean(axis=0)
medias_c2 = c2_treino[:, :-1].mean(axis=0)
medias_c3 = c3_treino[:, :-1].mean(axis=0)

cov_c1 = np.cov(c1_treino[:, :-1].T)
cov_c2 = np.cov(c2_treino[:, :-1].T)
cov_c3 = np.cov(c3_treino[:, :-1].T)
```

# Discriminante Quadrático Gaussiano

```python
def prob_dqg(media, cov, p_classe):
    
    m = cov.shape[0]
    pt_constante = p_classe/(sqrt(np.linalg.det(cov))*(2*math.pi)**(m/2.)) 
    
    def p(x):
        
        x = x.T
        return pt_constante*exp(-0.5*np.dot(np.dot((x - media).T, np.linalg.inv(cov)), (x - media)))
    
    return p
    

def dqg(medias, covs, ps_classe):
    
    funcoes = []
    
    for media, cov, p_classe in zip(medias, covs, ps_classe):
        
        funcoes.append(prob_dqg(media, cov, p_classe))
        
    def p(x):
        
        probabilidades = np.array([f(x) for f in funcoes])
        return probabilidades.argmax()
    
    return p
```

```python
p_dqg = dqg([medias_c1, medias_c2, medias_c3], [cov_c1, cov_c2, cov_c3], [1./3., 1./3., 1./3.])
```

## Naïve Bayes

```python
def prob_nb(media, cov, p_classe):
    
    var = cov.diagonal()
    
    pt_constante = 1./(np.sqrt(var)*sqrt(2*math.pi))
    
    def p(x):
        
        ps = pt_constante*np.exp(-1.*((x - media)**2)/(2*var))
        prod = np.prod(ps, axis=0)
        
        return p_classe*prod
    
    return p

def naive_bayes(medias, covs, ps_classe):
    
    funcoes = []
    
    for media, cov, p_classe in zip(medias, covs, ps_classe):
        
        funcoes.append(prob_nb(media, cov, p_classe))
        
    def p(x):
        
        probabilidades = np.array([f(x) for f in funcoes])

        return probabilidades.argmax()
    
    return p
```

```python
p_nb = naive_bayes([medias_c1, medias_c2, medias_c3], [cov_c1, cov_c2, cov_c3], [1./3., 1./3., 1./3.])
```

# Matriz de confusão

```python
def matriz_confusao(algoritmo, dados_teste, classes):
    
    m = len(classes)
    
    matriz = np.zeros((m,m))
    
    for dado_teste in dados_teste:
        
        # no arquivo vai de 1 a 3, uso 0 a 2
        esperado = dado_teste[-1] - 1
        
        predicao = algoritmo(dado_teste[:-1])
        
        matriz[esperado, predicao] += 1
        
    return matriz    
```

```python
dados_teste = np.vstack((c1_teste, c2_teste, c3_teste))
m = matriz_confusao(p_dqg, dados_teste, [0, 1, 2])
m_dqg = pd.DataFrame(m, columns=COLUMNS, index=COLUMNS)

m = matriz_confusao(p_nb, dados_teste, [0, 1, 2])
m_nb = pd.DataFrame(m, columns=COLUMNS, index=COLUMNS)
```
