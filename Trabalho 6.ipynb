```python
import os
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as pl
from sklearn import datasets
from sklearn.cross_validation import StratifiedShuffleSplit
from sklearn import metrics
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
iris = datasets.load_iris()
# cálculo dos labels
labels = np.unique(iris.target)

# sorteio dos dados utilizando quantidades de dados por classe iguais, com 3 iterações de sorteio
sss = StratifiedShuffleSplit(iris.target, n_iter=3, test_size=0.25)
```

## Kmeans

```python
# execução do kmeans com 2, 3, 4 e 5 clusters
ns_cluster = [2, 3, 4, 5]
models = {n:[] for n in ns_cluster}

for n_cluster in ns_cluster:
    model = KMeans(n_clusters=n_cluster)

    for train, test in sss:

        model.fit(iris.data[train, :], iris.target[train])
        
        y = iris.target[test]
        y_ = model.predict(iris.data[test, :])
        
        models[n_cluster].append((model, 
                                  {'acc':metrics.accuracy_score(y, y_),
                                   'cm': metrics.confusion_matrix(y, y_)
                                   })
                                 )
```

### Acurácia dos modelos gerados

```python
for key, model in models.iteritems():
    
    print 'número de clusters %d accuracy = [%.2f, %.2f, %.2f]' % (key, model[0][1]['acc'], model[1][1]['acc'], model[2][1]['acc'])
```

### Matriz de confusão para os modelos gerados

```python
fig, axes = pl.subplots(4, 3, figsize=(8, 10))

ticks = []
for i, n in enumerate(models):
    
    for j, res in enumerate(models[n]):
        
        cm = res[1]['cm']
        ticks.extend(cm.ravel().tolist())
        
        im = axes[i][j].matshow(res[1]['cm'])
        axes[i][j].set_title('n_clusters %d' % res[0].get_params()['n_clusters'])
        axes[i][j].set_xticks([])
        axes[i][j].set_yticks([])

valores = np.hstack(([0], np.array(list(set(ticks)))+1))
boundaries = valores-.5
fig.colorbar(im, ax=axes.ravel().tolist(), ticks=valores, drawedges=True, boundaries=boundaries)
```

## PCA

```python
pca = PCA(n_components=2)
X2d = pca.fit_transform(iris.data)
```
