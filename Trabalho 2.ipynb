# Função para plotar o gráfico com duas dimensões

```python
def plot2d(X, Y, color=None, filename=None):
    from matplotlib import pyplot as pl
        
    line = pl.plot(X, Y)
    if not color is None:
        pyplot.setp(line, color=color)
    
    if not filename is None:
        pyplot.savefig(filename)
        pyplot.clf()

def scatter2d(X, Y):
    from matplotlib import pyplot as pl
    
    pl.scatter(X, Y)
```

# Questão 1

```python
q1_dados = pd.read_csv(Q1_FILEPATH, sep=',', names=Q1_COLUMNS)
```

#### Função para adicionar uma coluna com ones à esquerda

```python
def add_indep(dados):
    ci = pd.DataFrame(dados)
    old_columns = dados.columns.tolist()[:]
    new_column_name = 'indep'
    new_columns = [new_column_name]
    new_columns.extend(old_columns)
    ci[new_column_name] = 1
    ci = ci[new_columns]
    return ci
```

```python
q1_dados = add_indep(q1_dados)
```

#### Partição dos dados em training-set e test-set

```python
q1_dados_treino = q1_dados[:Q1_TS_SIZE]
q1_dados_teste = q1_dados[Q1_TS_SIZE:]
```

#### Gradiente descendente estocástico

```python
def model_reg_log(dados, pesos):

    return 1./(1. + np.exp(-1*np.dot(dados[:-1], pesos)))

def bulk_model_reg_log(dados, pesos):
    
    return 1./(1. + np.exp(-1*np.dot(dados[dados.columns[:-1]], pesos)))
```

```python
def eqm(gerado, esperado):
    
    return np.sum((gerado - esperado)**2)

def reg_log(dados_treino, dados_teste, alfa, epocas):
    
    # num samples
    n = len(dados_treino.index)
    
    # cria o cycle dos dados - FAZER MELHOR!!!!! TALVEZ MUITO INEFICIENTE
    dados_cycle = cycle([np.array(dados_treino.ix[i]) for i in xrange(n)])
    
    # gerador de épocas
    l_epocas = xrange(epocas)
    
    # número de features(número de colunas - 2, do y e do w0)
    num_features = len(dados_treino.columns) - 2
    
    # pesos inicializados randomicamente
    pesos = np.random.rand(num_features + 1)
    
    # eqms
    eqms = []
    
    for epoca, row in zip(l_epocas, dados_cycle):
        y = model_reg_log(row, pesos)
        
        # calcula o erro nos dados - CALCULAR
        y_gerados = bulk_model_reg_log(dados_teste, pesos)
        erro_total = eqm(y_gerados, dados_teste[dados_teste.columns[-1]])
        eqms.append(erro_total)
        
        erro_iteracao = y - row[-1]
        
        # atualiza os pesos
        pesos = pesos - alfa*erro_iteracao*row[:-1]
        
        
    return pesos, eqms
        
    
```

```python
pesos, eqms = reg_log(q1_dados_treino, q1_dados_teste, Q1_ALFA, Q1_EPOCAS)
```

#### Gráfico época x erro (no test-set)
