#### Função para adicionar uma coluna com ones à esquerda

```python
def add_indep(dados):
    
    ci = pd.DataFrame(dados)
    old_columns = dados.columns.tolist()[:]
    new_column_name = 'indep'
    new_columns = [new_column_name]
    new_columns.extend(old_columns)
    ci[new_column_name] = 1
    ci = ci[new_columns]
    
    return ci
```

#### Função mapFeature

```python
# função que cria, a partir de um data frame com 4 colunas(indep, feature 1, feature 2, classe) um data frame com colunas 
#  com todos resultados de combinações das colunas de features em polinômios
#  >> entrada DataFrame e expoente máximo
def map_feature(dados, n):
    
    if n == 1:
        return dados
    
    # expoentes
    expoentes = ((i, j) for i in xrange(1, n+1) for j in xrange(i+1))
    
    resultado = pd.DataFrame(index=dados.index)
    resultado = add_indep(resultado)
    
    for expoente in expoentes:
        
        i = expoente[0]
        j = expoente[1]
        
        resultado['feature (%d, %d)' % (i-j, j)] = (dados[dados.columns[1]]**(i-j)) * dados[dados.columns[2]]**j
        
    resultado[dados.columns[-1]] = dados[dados.columns[-1]] # copia a coluna da classe
    
    return resultado

# versão recebendo matriz x e y
def map_feature_matrix(x, y, n):
    
    expoentes = [(i, j) for i in xrange(1, n+1) for j in xrange(i+1)]
    
    resultado = np.ndarray((x.shape[0], x.shape[1], len(expoentes)+1))
    
    for i in xrange(x.shape[0]):
        for j in xrange(x.shape[1]):
            
            # independente
            resultado[i,j,0] = 1.
            for k, expoente in enumerate(expoentes):
                resultado[i,j,k+1] = (x[i,j]**(expoente[0]-expoente[1]))*y[i,j]**expoente[1]
                
    
    return resultado            
```

```python
def model_reg_log(dados, pesos):
    return 1./(1. + np.exp(-1.*np.dot(dados[:-1], pesos)))

def bulk_model_reg_log(dados, pesos):
    return 1./(1. + np.exp(-1.*np.dot(dados[dados.columns[:-1]], pesos)))

def bbulk_model_reg_log(Z, pesos):
    
    resultado = np.ndarray(Z.shape[:-1])
    
    for i in xrange(Z.shape[0]):
        for j in xrange(Z.shape[1]):
            
            xs = Z[i,j]
            
            resultado[i,j] = 1./(1. + np.exp(-1.*np.dot(xs, pesos)))
            
    return resultado
```

#### Plot dos dados

```python
# produz o scatter dos dados da segunda e terceira coluna, para as duas classes na quarta coluna
def q1_scatter(dados, pesos=None, c_min=.5, c_max=3.5):
    
    # sem pesos, scatter apenas dos dados
    if pesos == None:
        
        pl.scatter(dados[dados['is_aprovado']==1][dados.columns[1]], dados[dados['is_aprovado']==1][dados.columns[2]], marker='x', c='b')
        pl.scatter(dados[dados['is_aprovado']==0][dados.columns[1]], dados[dados['is_aprovado']==0][dados.columns[2]], marker='o', c='r')
    
    else:
        # com pesos, scatter dos dados e das predições
        r = bulk_model_reg_log(dados, pesos)
        f, (fig1, fig2) = pl.subplots(2, sharex=True, sharey=True)
        f.set_size_inches(10, 10)
        
        fig1.scatter(dados[dados['is_aprovado']==1][dados.columns[1]], dados[dados['is_aprovado']==1][dados.columns[2]], marker='x', c='b')
        fig1.scatter(dados[dados['is_aprovado']==0][dados.columns[1]], dados[dados['is_aprovado']==0][dados.columns[2]], marker='o', c='r')
        
        fig2.scatter(dados[r>=.5][dados.columns[1]], dados[r>=.5][dados.columns[2]], marker='x', c='b')
        fig2.scatter(dados[r<.5][dados.columns[1]], dados[r<.5][dados.columns[2]], marker='o', c='r')

        # contour
        u = np.linspace(c_min, c_max, 50)
        v = np.linspace(c_min, c_max, 50)
        
        U,V = np.meshgrid(u,v)
        
        Z = map_feature_matrix(U, V, 1)
        
        Zs = bbulk_model_reg_log(Z, pesos)
        Zs[Zs>=.5] = 1
        Zs[Zs<.5] = 0

        fig2.contourf(u, v, Zs, 1, alpha=.2)
        
    
    
q1_scatter(q1_dados)
```

#### Gradiente descendente estocástico

```python
def eqm(gerado, esperado):
    
    return np.sum((gerado - esperado)**2)/len(gerado)

PATH_IMAGES = 'reg_log_img'
# retorna um DataFrame com os pesos, eqms x dados de treino, eqms x dados de teste, para todas as épocas calculadas
def reg_log(dados_treino, dados_teste, alfa, epocas, save=False):
    
    # num samples
    n = len(dados_treino.index)
    
    # cria o cycle dos dados - FAZER MELHOR!!!!! TALVEZ MUITO INEFICIENTE
    dados_cycle = cycle([np.array(dados_treino.ix[i]) for i in dados_treino.index])
    
    # gerador de épocas
    l_epocas = xrange(epocas)
    
    # número de features(número de colunas - 2, do y e do w0)
    num_features = len(dados_treino.columns) - 2
    
    # pesos inicializados randomicamente
    pesos = np.random.uniform(-1.0, 1.0, num_features + 1)
    
    # eqms
    eqms = []
    
    for epoca, row in zip(l_epocas, dados_cycle):
       
        y = model_reg_log(row, pesos)
        
        # calcula o erro nos dados - CALCULAR
        y_gerados_treino = bulk_model_reg_log(dados_treino, pesos)
        erro_total_treino = eqm(y_gerados_treino, dados_treino['is_aprovado'])
        
        y_gerados_teste = bulk_model_reg_log(dados_teste, pesos)
        erro_total_teste = eqm(y_gerados_teste, dados_teste['is_aprovado'])
        
        eqms.append((pesos, erro_total_treino, erro_total_teste))
        
        erro_iteracao = y - row[-1]
        
        # atualiza os pesos
        pesos = pesos - alfa*erro_iteracao*row[:-1]
        
        if save:
            q1_scatter(pesos)
            pl.savefig(os.path.join(PATH_IMAGES, 'img%04d.png' % epoca))
            pl.clf()
            pl.close()
        
    return pd.DataFrame(eqms, columns=['pesos', 'eqms treino', 'eqms teste'])
```

```python
eqms = reg_log(q1_dados_treino_norm, q1_dados_teste_norm, Q1_ALFA, Q1_EPOCAS+1000)
```

#### k-fold

```python
# gera uma lista de índices de partição
# por exemplo, size=20 e k=10 resultado=[(0:10), (10:20)]
def indices_particao(size, k):
    
    i_s = np.arange(0, size, k)
    i_ss = zip(i_s[:-1], i_s[1:])
    i_ss.append((i_ss[-1][1], size))
    
    return i_ss
```

```python
def k_fold(dados, k, alfa, epocas):
    
    # número de samples
    n = len(dados.index)
    
    # indices do particionamento
    is_part = indices_particao(n, k)
    
    # resultados
    resultados = []
    
    for i_part in is_part:
        
        dados_teste = dados[i_part[0]: i_part[1]]
        dados_treino = dados.drop(dados.index[i_part[0]: i_part[1]])
        
        eqms = reg_log(dados_treino, dados_teste, alfa, epocas)
        
        # calcula o peso com menor eqms de treino
        eqms_minimos_treino = eqms[eqms['eqms treino'] == np.min(eqms['eqms treino'])]
        eqms_minimo_treino = eqms_minimos_treino.ix[eqms_minimos_treino.index[0]]
        
        # calcula o peso com menor eqms de teste
        eqms_minimos_teste = eqms[eqms['eqms teste'] == np.min(eqms['eqms teste'])]
        eqms_minimo_teste = eqms_minimos_teste.ix[eqms_minimos_teste.index[0]]
        
        resultados.append((eqms_minimo_treino['pesos'], eqms_minimo_treino['eqms treino'], eqms_minimo_teste['pesos'], eqms_minimo_teste['eqms teste']))
        
    return pd.DataFrame(resultados, columns=['pesos treino', 'eqms treino', 'pesos teste', 'eqms teste'])
        
    
```
