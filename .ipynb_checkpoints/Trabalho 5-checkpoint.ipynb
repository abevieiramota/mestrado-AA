```python
import numpy as np
import matplotlib.pyplot as pl
import scipy.io as io
import os as os
from sklearn.svm import SVC
%matplotlib inline
```

# Constantes

```python
# pasta com os dados
PATH_DADOS = "Dados Exercicio 5"

# constantes da questão 1

Q1_FILEPATH = os.path.join(PATH_DADOS, "ex5data1.mat")
Q2_FILEPATH = os.path.join(PATH_DADOS, "ex5data2.mat")
```

# Carregamento dos dados

```python
q1_dados = io.loadmat(Q1_FILEPATH)
X = np.array(q1_dados['X'])
Y = np.array(q1_dados['y'])[:, 0]

q2_dados = io.loadmat(Q2_FILEPATH)
X2 = np.array(q2_dados['X'])
Y2 = np.array(q2_dados['y'])[:, 0]
```

```python
def scatter_dados(X, Y):
    pl.scatter(X[Y==0][:, 0], X[Y==0][:, 1], c='r', label='C0', s=40)
    pl.scatter(X[Y==1][:, 0], X[Y==1][:, 1], c='b', label='C1', s=40, marker='x')
    pl.legend(loc='lower right')
    
scatter_dados(X, Y)
```

# LinearSVM

```python
model_1 = SVC(kernel='linear', C=1.)
model_1.fit(X, Y)

model_100 = SVC(kernel='linear', C=100.)
model_100.fit(X, Y)

model_001 = SVC(kernel='linear', C=.001)
model_001.fit(X, Y)
```

```python
def boundary(model, X, Y, inc=1., alpha=.2, cmap='gray', step=.2):
    
    x1_min, x1_max = X[:, 0].min()-inc, X[:, 0].max()+inc
    x2_min, x2_max = X[:, 1].min()-inc, X[:, 1].max()+inc
    
    xx, yy = np.meshgrid(np.arange(x1_min, x1_max, step),
                     np.arange(x2_min, x2_max, step))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)

    pl.contourf(xx, yy, Z, alpha=alpha, cmap=cmap)
```

# Kernel Linear comm C=1.0

```python
boundary(model_1, X, Y)
scatter_dados(X, Y)
```

# Kernel Linear com C=100.0

```python
boundary(model_100, X, Y)
scatter_dados(X, Y)
```

# $\sigma$ = 0.1

```python
sigma = 0.1
gamma = 1./(2*sigma**2)
model_rbf = SVC(gamma=gamma, C=1.0)
model_rbf.fit(X2, Y2)

boundary(model_rbf, X2, Y2, inc=.001, alpha=.5, cmap='gray', step=.01)
scatter_dados(X2, Y2)
```

# $\sigma$ = 0.2

```python
sigma = 0.2
gamma = 1./(2*sigma**2)
model_rbf = SVC(gamma=gamma, C=1.0)
model_rbf.fit(X2, Y2)

boundary(model_rbf, X2, Y2, inc=.001, alpha=.5, cmap='gray', step=.01)
scatter_dados(X2, Y2)
```
